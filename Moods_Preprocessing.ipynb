```python
#importing the required libraries
import pandas as pd

#loading the Excel file after converting it online
df = pd.read_excel('mood_data.xlsx',  parse_dates=True, squeeze=True)

#Checking the data
df
```

```python
#Dropping the not required columns and renaming the ones which are named oddly.
df = df.drop(columns=['entries__type', 'entries__notes', 'entries__isHighlighted'])
df = df.rename(columns={"entries__value": "rating"})

#Checking the data again
df.head(10)
```

```python
#Getting the rolling average of the mood ratings
df['roll'] = df.rolling(window=5).mean()
```

```python
#Checking the data again with rolling averages added
df
```

```python
#Creating the columns for months and days
df['date'] = pd.to_datetime(df['date'])
df['day'] = df['date'].dt.day_name()
df['month'] = df['date'].dt.month
```

```python
#Creating a function to convert ratings into a certain mood 
def change_func(x):
    if x == 1:
        return "Sad"
    elif x == 2:
        return "Unhappy"
    elif x == 3:
        return "Okay"
    elif x == 4:
        return "Happy"
    elif x == 5:
        return "Very Happy"

df["mood"] = df["rating"].apply(lambda x: change_func(x))    

#Creating a function to convert month value into month names
def change_month_func(x):
    if x == 1:
        return "January"
    elif x == 2:
        return "February"
    elif x == 3:
        return "March"
    elif x == 4:
        return "April"
    elif x == 5:
        return "May"
    elif x == 6:
        return "June"
    elif x == 7:
        return "July"
    elif x == 8:
        return "August"
    elif x == 9:
        return "September"
    elif x == 10:
        return "October"
    elif x == 11:
        return "November"
    elif x == 12:
        return "December"

df["month"] = df["month"].apply(lambda x: change_month_func(x))
```

```python
#Chekcing our final data
df
```

```python
#Exporting to Execel for visualizations
df.to_excel("mood_data_cleaned.xlsx")  
```
